```python
#importing my libraries 
import numpy as np 
import csv
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
#defining and reading the csv files 
df = pd.read_csv('VHT.csv')
df1 = pd.read_csv('UScovidcases.csv')
```

```python
#setting the start and end dates for my graphs 
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2023-4-30')

#condensing the date portion of the csv files to fit well into the graph 

df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
stock_data = df1.set_index('Date')
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(stock_data["Close"])

#setting the titlte, y axis label, and size of the graph

plt.title('VHT etf price from 2020 to 2023')
plt.ylabel('price by the 10s')
plt.gcf().set_size_inches(12,8)
```

```python
#same thing done for this portion for the covid cases 
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2023-4-30')                         
df1['date'] = pd.to_datetime(df1['date']) 
new_df1 = (df1['date']>= start_date) & (df1['date']<= end_date)
df2 = df.loc[new_df1]
covid_data = df1.set_index('date')
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(covid_data["cases"])
plt.title('Covid Cases in the US from 2020 to 2023')
plt.ylabel('Cases by the 100 millions')
plt.gcf().set_size_inches(12,8)
```
